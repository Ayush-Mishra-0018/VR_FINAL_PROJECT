```python
import pandas as pd
import os

df = pd.read_csv("./results/memberB_outputs/captions.csv")

def parse_filename(image_name):
    parts = image_name.replace(".jpg", "").split("__")
    # parts: [crop_class, img, GENDER, CATEGORY, item_id, shot]
    try:
        return pd.Series({
            "gender":   parts[2],
            "category": parts[3],
            "item_id":  parts[4],
            "shot":     parts[5],
        })
    except IndexError:
        return pd.Series({
            "gender": "", "category": "", "item_id": "", "shot": ""
        })

parsed = df["image_name"].apply(parse_filename)
metadata = pd.concat([df, parsed], axis=1)

# reorder columns cleanly for C
metadata = metadata[[
    "original_image",
    "image_name",
    "crop_class",
    "gender",
    "category",
    "item_id",
    "shot",
    "caption"
]]

print(f"Total rows: {len(metadata)}")
print(f"Unique item_ids: {metadata['item_id'].nunique()}")
print(f"Nulls:\n{metadata.isnull().sum()}")
print(f"\nSample:\n{metadata.head(5).to_string()}")

metadata.to_csv("./results/memberB_outputs/metadata.csv", index=False)
print("Saved: metadata.csv")
```
